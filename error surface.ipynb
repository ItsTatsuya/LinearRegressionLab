```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('lr-data.csv')

m_values = np.arange(-10, 10, 0.1)
c_values = np.arange(-10, 10, 0.1)

x_values = data.iloc[:, 0].values
y_values = data.iloc[:, 1].values


mc_set = list(zip(m_values, c_values))
errors = []

for m, c in mc_set:

    predicted_y = m * x_values + c

    error = np.sum((y_values - predicted_y) ** 2)
    errors.append(error)


fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(m_values, c_values, errors)

ax.set_xlabel('m')
ax.set_ylabel('c')
ax.set_zlabel('Error')
plt.title('Error Surface')
plt.show()
```
