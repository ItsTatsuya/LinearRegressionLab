```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from itertools import product
from mpl_toolkits.mplot3d import Axes3D
```

```python
# Load your data
data = pd.read_csv('lr-data.csv')

# Generate m and c values
m_values = np.arange(-10, 10, 0.1)
c_values = np.arange(-10, 10, 0.1)

x_values = data.iloc[:, 0].values
y_values = data.iloc[:, 1].values

# Create meshgrid for m and c values
M, C = np.meshgrid(m_values, c_values)

# Initialize an array for errors
errors = np.zeros(M.shape)

# Calculate errors
for i, (m, c) in enumerate(product(m_values, c_values)):
    predicted_y = m * x_values + c
    errors.flat[i] = np.sum(np.abs(y_values - predicted_y))

# Create a 3D surface plot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Plot the surface
ax.plot_surface(M, C, errors, cmap='viridis')

# Set labels
ax.set_xlabel('m')
ax.set_ylabel('c')
ax.set_zlabel('Error')
plt.title('Error Surface')
plt.show()
```
